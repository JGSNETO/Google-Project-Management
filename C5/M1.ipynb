## Introduction to Agile

## A brief history of Agile

- *Waterfal*: The sequential or linear ordering of phases
- *Agile*:
1. Being able to move quickly and easily 
2. Flexibility and the willingness and ability to change and adapt
3. Iteractive approach(Sprint)


- Agile´s iterative approach enables a project to move quickly, as well as making it adaptive to change. 
- *Agile project management*: An approach to project and team management that embodies "agility" and is based on the Agile Manifesto. The manifesto is a collection of *four values* and *12 principles* that define the mindiset that all agile teams should trive for. 
- Agile values, principles, and frameworks have been applied to every industry. 
- Agile methods draw heavily on Lean manufacturing principles.

## Distinguishing Agile from Waterfall

- Agile was created in response to the strict linear process of Waterfall
- Agile embraces the reality that the world, customer markets, and users are unvertain and unpredictable. 
- Agile aims to get customer feedback more quickly
- Working with an Agile mindset means always finding ways to work more efficiently by focusing on streamlining the process without reducing product quality or value. 
- Agile aims to reduce waste. 
- *Aspects of a project*
1. Requirements: Conditions that must be met or tasks that must be finished to ensure the successful completion of the project. 
2. Documentation: In Agile there is an empahis on real-time person-to-person conversations. 
3. Deliverables

## The four value of the Agile Manifesto. 
- We are uncovering better ways of developing software by doing it and helping others do it. Through this work we have come to value:

1. Individuals and interactions *over* processes and tools
2. Working software *over* comprehensive documentation
3. Customer collaboration *over* contract negotiation
4. Responding to change *over* following a plan
